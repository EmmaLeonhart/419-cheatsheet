## Section 6: Linear Regression with NumPy

In this section, we'll apply what you've learned about NumPy to implement a basic linear regression model and calculate its gradients.

**Task 6.1: Create Dummy Data**

Let's create some synthetic data for regression. We'll generate 100 data points:
- A single input feature `X` generated randomly using `np.random.rand` in the range [0, 1].
- A corresponding target variable `y`.  Let's create it using a linear model: `y = 2*X + 1 + noise` where `noise` is random Gaussian noise with `mean=0` and small `std=0.2` generated using `np.random.normal`.  Reshape both X and y to be column vectors

```python
X = np.random.rand(100, 1)

noise = np.random.normal(0, 0.2, size=(100, 1))

y = 2 * X + 1 + noise

print("X shape:", X.shape)
print("y shape:", y.shape)
```

**Task 6.2: Linear Model Prediction**

Define a function `linear_model(X, W, b)` to compute the predictions of a linear model. The linear model will be: `y_hat = W*X + b` where `W` is the weight and `b` is the bias.

Use it to compute initial predictions using a randomly initialized weight and bias. Use `np.random.randn` to initialize the weights and bias.

```python
def linear_model(X, W, b):
    return W * X + b

W = np.random.randn(1)
b = np.random.randn(1)

y_hat = linear_model(X, W, b)

print("W:", W)
print("b:", b)
print(y_hat)
```

**Task 6.3: Loss Function: Mean Squared Error**

Define a function `mean_squared_error(y_hat, y)` that calculates the mean squared error (MSE) between the predicted values (`y_hat`) and the actual target values (`y`). The MSE is calculated as:

`MSE = 1/N * sum((y_hat - y)^2)` where `N` is the number of samples.
Compute the MSE with the initial predictions.

```python
def mean_squared_error(y_hat, y):
    N = y.shape[0]
    return (1 / N) * np.sum((y_hat - y) ** 2)

mse = mean_squared_error(y_hat, y)

print("MSE:", mse)
```

**Task 6.4: Gradient Calculation**

Compute the gradients of the MSE loss with respect to the weight `W` and bias `b`.
The gradients are given by:
`dMSE/dW = 2/N * sum((y_hat - y) * X)`
`dMSE/db = 2/N * sum(y_hat - y)`

Implement functions `gradient_w(X, y_hat, y)` to calculate gradient of MSE w.r.t to weight, and `gradient_b(y_hat, y)` to calculate gradient of MSE w.r.t to bias.

Compute these gradients using the initial predictions.

```python

def gradient_w(X, y_hat, y):
    N = y.shape[0]
    return (2 / N) * np.sum((y_hat - y) * X)

# Gradient w.r.t. bias b
def gradient_b(y_hat, y):
    N = y.shape[0]
    return (2 / N) * np.sum(y_hat - y)


grad_W = gradient_w(X, y_hat, y)
grad_b = gradient_b(y_hat, y)

print("Gradient w.r.t W:", grad_W)
print("Gradient w.r.t b:", grad_b)
```

**Task 6.5: Gradient Descent Update**

1. Set a learning rate to a small value such as 0.01
2. Update the weight and bias using gradient descent:
   `W = W - learning_rate * dMSE/dW`
   `b = b - learning_rate * dMSE/db`
3. Compute the loss and gradients again with the updated parameters. What happened to the loss?

```python

learning_rate = 0.01


W = W - learning_rate * grad_W
b = b - learning_rate * grad_b

y_hat_new = linear_model(X, W, b)


mse_new = mean_squared_error(y_hat_new, y)


grad_W_new = gradient_w(X, y_hat_new, y)
grad_b_new = gradient_b(y_hat_new, y)

print("Updated W:", W)
print("Updated b:", b)

print("New MSE:", mse_new)
print("New gradient w.r.t W:", grad_W_new)
print("New gradient w.r.t b:", grad_b_new)
```

**Task 6.6: Training Loop**

Create a loop that iterates for 100 or more iterations and updates the parameters for each iteration. Plot the training loss as a function of the iterations.

```python
learning_rate = 0.01
num_iterations = 100

losses = []

for i in range(num_iterations):

    y_hat = linear_model(X, W, b)
    

    mse = mean_squared_error(y_hat, y)
    losses.append(mse)
    

    grad_W = gradient_w(X, y_hat, y)
    grad_b = gradient_b(y_hat, y)
    

    W = W - learning_rate * grad_W
    b = b - learning_rate * grad_b

import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error")
plt.title("Training Loss vs Iteration")
plt.show()

```
